```python
import pandas as pd

# Загрузка данных из файла CSV
df = pd.read_csv("C:\\Users\\swer\\orders.csv")

# 1. Количество уникальных городов
unique_cities = df['city_id'].nunique()

# 2. Количество ресторанов, специализирующихся на рыбе
fish_restaurants = df[df['spec'] == 'рыба']['vendor_id'].nunique()

# 3. Количество колонок типа данных float64
float_columns = len(df.select_dtypes(include=['float64']).columns)

# 4. Количество дней у ресторана с идентификатором 40065, где успешных заказов было менее 20
days_less_than_20_orders = df[(df['vendor_id'] == 40065) & (df['successful_orders'] < 20)].shape[0]

# Вывод ответов
print(unique_cities, fish_restaurants, float_columns, days_less_than_20_orders, sep=', ')
```

```python
# Предположим, что ожидаемые результаты равны 4, 0, 2, 45
assert unique_cities == 4
assert fish_restaurants == 0
assert float_columns == 2
assert days_less_than_20_orders == 45
```
