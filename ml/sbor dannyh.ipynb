```python
import requests
import pandas as pd

# Функция для сбора данных с веб-сайта
def scrape_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Проверка на ошибки HTTP

        # Проверка на успешный статус код
        if response.status_code == 200:
            data = response.json()  # Предположим, что данные получены в формате JSON
            return data
        else:
            print(f"Ошибка: Не удалось получить данные. Статус код: {response.status_code}")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Ошибка при выполнении запроса: {e}")
        return None
    except ValueError as e:
        print(f"Ошибка при преобразовании данных в JSON: {e}")
        return None

# URL для сбора данных
url = "https://api.example.com/data"

# Сбор данных
data = scrape_data(url)

if data:
    # Преобразование данных в DataFrame
    df = pd.DataFrame(data)

    # Сохранение данных в CSV файл
    df.to_csv('dataset.csv', index=False)
    print("Данные успешно сохранены в файл dataset.csv")
```

```python
import requests
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Необходимо скачать пакет токенизатора, если он еще не установлен
nltk.download('punkt')

def scrape_data(url):
    """Функция для сбора данных с веб-сайта в формате JSON."""
    try:
        response = requests.get(url)
        response.raise_for_status()  # Автоматическая проверка на ошибки HTTP
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при выполнении запроса: {e}")
    except ValueError:
        print("Ошибка при преобразовании данных в JSON")
    return None

def tokenize_text(data_frame, text_column):
    """Функция для токенизации текстовых данных в DataFrame."""
    data_frame['tokenized_text'] = data_frame[text_column].apply(word_tokenize)

def main():
    url = "https://api.example.com/data"
    data = scrape_data(url)
    if data:
        df = pd.DataFrame(data)
        if 'text_column' in df.columns:
            tokenize_text(df, 'text_column')
            df.to_csv('tokenized_dataset.csv', index=False)
            print("Данные успешно сохранены в файл tokenized_dataset.csv")
        else:
            print("Ошибка: Колонка с текстом 'text_column' отсутствует в данных.")
    else:
        print("Данные не были получены.")

if __name__ == "__main__":
    main()
```
